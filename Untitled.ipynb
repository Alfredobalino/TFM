```python

import pandas as pd
text_file = open("BD_objGeograf.txt",encoding='cp1252') ##Revisar el encoding, las letras con tilde las saca mal
lines = text_file.readlines()

d = "}"
for line in lines:
    result =  [e+d for e in line.split(d) if e] 

#result = lines[0].split('}{')
#result[len(result)-1] = result[len(result)-1][:-1]
result
```

```python

import ast
list=[]

i = 0
t=len(result)
while i < t:
    try:
        datos = ast.literal_eval(result[i])
        frame = pd.DataFrame(datos, index = [0])
        list.append(frame)
        resultado = pd.concat(list)
    except Exception as e:
        s = str(e)
        print(s)
    i += 1 
    
        
resultado#["rings"]
```

```python
import csv 
resultado.to_csv('BD_Objgeo.csv', sep='\t', encoding='utf-8', index = False)
```

```python
import MySQLdb

# creo la conexión
mydb = MySQLdb.connect(host='localhost',
    user='root',
    passwd='',
    db='test')
cursor = mydb.cursor()

# creo la sentencio
insert_stmt = 'INSERT INTO testcsv (OBJECTID, RINGS) VALUES("%s", "%s")'

with open('BD_Objgeo.csv', encoding="utf8") as csvfile:
    csv_data = csv.reader(csvfile, delimiter='\t')
    for row in csv_data:

        cursor.execute(insert_stmt, row)

#close the connection to the database.
mydb.commit()
cursor.close()
print ("Done")
```
