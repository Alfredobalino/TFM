```python
import pandas as pd
text_file = open("BD_atributos.txt",encoding='850') ##Revisar el encoding, las letras con tilde las saca mal
lines = text_file.readlines()


result = lines[0].split('][')
result[len(result)-1] = result[len(result)-1][:-1]
result
```

```python
import ast
list=[]

i = 1
while i < len(result):
    datos = ast.literal_eval(result[i])
    frame = pd.DataFrame(datos, index = [0])
    list.append(frame)
    i += 1    
    
resultado = pd.concat(list)
resultado


```

```python
import csv 
resultado.to_csv('BD_atributos.csv', sep='\t', encoding='UTF-8', index = False)
```
